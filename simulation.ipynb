```python
#5년치 기대 승률

import random
import numpy as np
import pandas as pd

total_rank = [['doosan'],['hanwha'],['kia'],['kiwoom'],['kt'],['lg'],['lotte'],['nc'],['samsung'],['ssg']]
rank_count = [['doosan'],['hanwha'],['kia'],['kiwoom'],['kt'],['lg'],['lotte'],['nc'],['samsung'],['ssg']]
team = ['doosan','hanwha','kia','kiwoom','kt','lg','lotte','nc','samsung','ssg']
def simulation():
    #팀 = [기대 승률]
    ssg = [0.502]
    lg = [0.526]
    doosan = [0.597]
    lotte = [0.472]
    kiwoom = [0.548]
    samsung = [0.464]
    kt = [0.502]
    kia = [0.493]
    hanwha = [0.415]
    nc = [0.516]
    game = [doosan, hanwha, kia, kiwoom, kt, lg, lotte, nc, samsung, ssg]
    team = ['doosan','hanwha','kia','kiwoom','kt','lg','lotte','nc','samsung','ssg']
    game_win =[]
    for i in range(10):
        for j in range(10):
            for x in range(16):
                if i>j:
                    a = game[i][0]
                    b = game[j][0]

                    win = random.choices(['a','b'], weights = (a, b))
                    if win == ['a']:
                        game[i].append(1)
                        game[j].append(0)
                    else:
                        game[i].append(0)
                        game[j].append(1)
    #시뮬레이션 종료 후 확률 제거
    for y in range(10):
        game[y].pop(0)

    for k in range(10):
        w = sum(game[k])
        game_win.append(w)

    #팀 이름과 승 수 리스트를 합쳐 순위 계산
    z = list(zip(team, game_win))
    win_loose =  sorted(z, key=lambda z: z[1], reverse = True)
    ranking = []
    for i in range(10):
        ranking.append(list(win_loose[i]))
        ranking[i].remove(ranking[i][1])
        ranking[i].append(i+1)
    ranking = sorted(ranking)
    
    for i in range(10):
        total_rank[i].append(ranking[i][1])
        

#시즌 반복 100번 시행
for i in range(10000):
    simulation()

#각 팀별 평균 등수

mean_rank = total_rank.copy()
for i in range(10):
    mean_rank[i].pop(0)
    mean_rank[i] = sum(mean_rank[i])/len(mean_rank[i])
mean_rank = list(zip(team, mean_rank))
print("각 팀별 평균 등수")
print(mean_rank)
df_mean = pd.DataFrame(mean_rank)
df_mean.to_excel('meanrank.xlsx', index=False)

#각 팀 등수별 횟수

for k in range(10):
    for j in range(10):
        count = 0
        for i in range(len(total_rank[0])):
            if total_rank[k][i] == j+1:
                count+=1
        rank_count[k].append(count)
print("각 팀 등수별 횟수")
print(rank_count)
df_count = pd.DataFrame(rank_count)
df_count.to_excel('rankcount(5years).xlsx', index=False)
```
